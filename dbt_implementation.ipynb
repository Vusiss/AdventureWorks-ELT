```python
import pandas as pd
import pymssql
```

```python
server = 'localhost' # Nazwa serwera MS SQL
database = 'AdventureWorks2014' # Nazwa bazy danych
user = 'sa' # Nazwa użytkownika
password = 'qwerty' # Hasło użytkownika
```

```python
try:
 conn = pymssql.connect(server, user, password, database)
 print("Połączenie z bazą danych MS SQL udane!")
except pymssql.Error as e:
 print(f"Błąd połączenia z bazą danych: {e}")
 exit() 
```

```python
cursor = conn.cursor()
```

```python
sql_query = 'SELECT TOP 10 * FROM Person.Person'
```

```python
try:
    cursor.execute(sql_query)
except pymssql.Error as e:
    print(f"Błąd wykonania zapytania SQL: {e}")
    conn.close() # Pamiętaj o zamknięciu połączenia w przypadku błędu
    exit()
```

```python
rows = cursor.fetchall()
for row in rows:
    print(row) # Wyświetl każdy wiersz (krotkę) 
```

```python
cursor.close()
conn.close() 
```

```python
from sqlalchemy import create_engine, text 
```

```python
username = "sa"
password = "qwerty"
server = "127.0.0.1"
port = 1433
database = "AdventureWorks2014"

connection_string = f"mssql+pymssql://{username}:{password}@{server}:{port}/{database}"
```

```python
try:
    engine = create_engine(connection_string)
    print("Połączenie z bazą danych MS SQL (SQLAlchemy) udane!")
except Exception as e: # SQLAlchemy może rzucać różnymi typami wyjątków
    print(f"Błąd połączenia z bazą danych (SQLAlchemy): {e}")
    exit() 
```

```python
sql_query = text("SELECT TOP 10 * FROM Person.Person") # Zapytanie SQL jako obiekt text()
try:
    with engine.connect() as connection: # Użyj 'with' aby automatycznie zamknąć połączenie
        result = connection.execute(sql_query) # Wykonanie zapytania
        rows = result.fetchall() # Pobranie wszystkich wierszy 
        for row in rows:
            print(row)
except Exception as e:
    print(f"Błąd wykonania zapytania SQL (SQLAlchemy): {e}")
    exit() 
```
